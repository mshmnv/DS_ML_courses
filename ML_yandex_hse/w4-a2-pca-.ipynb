# Составление фондового индекса
- Загрузите данные close_prices.csv. В этом файле приведены цены акций 30 компаний на закрытии торгов за каждый день периода. 
- На загруженных данных обучите преобразование PCA с числом компоненты равным 10. Скольких компонент хватит, чтобы объяснить 90% дисперсии?
- Примените построенное преобразование к исходным данным и возьмите значения первой компоненты. 
- Загрузите информацию об индексе Доу-Джонса из файла djia_index.csv. Чему равна корреляция Пирсона между первой компонентой и индексом Доу-Джонса? 
- Какая компания имеет наибольший вес в первой компоненте? Укажите ее название с большой буквы.

```python
import pandas as pd
import numpy as np      # numpy.corrcoef() - Pearson correlation coefficient
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('data/close_prices.csv')
X = data.iloc[:,1:]
X
```

```python
pca = PCA(n_components=10)
pca.fit(X)
pca.explained_variance_ratio_
```

**Сколько компонент хватит, чтобы объяснить 90 % дисперсии?**

```python
sum_ratio = 0
count = 0
for i in pca.explained_variance_ratio_:
    sum_ratio += i
    count += 1
    if sum_ratio > 0.9:
        break

print("We need {} components.".format(count))
```

**Примените построенное преобразование к исходным данным и возьмите значения первой компоненты.**

```python
first_comp = pca.transform(X)[:, 0]
first_comp
```

**Чему равна корреляция Пирсона между первой компонентой и индексом Доу-Джонса?**

```python
data2 = pd.read_csv('data/djia_index.csv')
data2 = data2.iloc[:,1]

np.corrcoef(first_comp, data2)[0,1]

# ans_file = open('answer_2.txt', 'w')
# print(np.corrcoef(first_comp, data2)[0,1], file=ans_file, end="")
# ans_file.close()
```

**Какая компания имеет наибольший вес в первой компоненте?**

```python
data.columns[np.argmax(pca.components_[0]) + 1]
```
